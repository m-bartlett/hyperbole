```python
def drawHyperDragon(drawing, depth, a1=0, a2=-180, **style):
    if depth <= 0:
        hl = Line.fromPoints(*Ideal.fromDegree(a1),
                             *Ideal.fromDegree(a2))
        drawing.draw(hl, **style)
    else:
        drawHyperDragon(drawing, depth-1, (a1+a2)/2, a2, **style)
        drawHyperDragon(drawing, depth-1, (a1+a2)/2+360, a1, **style)

d = Drawing(2.1, 2.1, origin='center')
d.draw(euclid.shapes.Circle(0, 0, 1),
       fill='none', stroke_width=0.006, stroke='black')
lineStyle = dict(hwidth=0.03, fill='black')
drawHyperDragon(d, 6, **lineStyle)

d.setRenderSize(w=400)
d.saveSvg('images/hyperDragon.svg')
d  # Renders badly in Chrome
```

```python
def hypPolyEdgeConstruct(p, q):
    pi, pi2 = math.pi, math.pi*2
    th = pi2/q
    phi = pi2/p
    ang1 = pi-phi/2-th/2-pi/2
    ang2 = th/2 + pi/2
    a = math.sin(ang2)/math.sin(ang1)
    b = math.sin(phi/2)/math.sin(ang1)
    rP = math.sqrt(1/(a**2-b**2))
    rC = a*rP
    rFromC = b*rP
    #return rC, rFromC
    t1 = pi - math.asin(rC / (rFromC / math.sin(phi/2)))
    t2 = pi - t1 - phi/2
    r = math.sin(t2) * (rFromC / math.sin(phi/2))
    return r
```

```python
class TreeNode:
    def __init__(self, point, children=None, parent=None):
        self.point = point
        self.children = [] if children is None else children
        self.parent = parent
    def addChild(self, child):
        child.parent = self
        self.children.append(child)
    def transform(self, trans):
        self.point = trans.applyToPoint(self.point)
        for child in self.children:
            child.transform(trans)
    def transformed(self, trans, newParent):
        newPoint = trans.applyToPoint(self.point)
        newNode = TreeNode(newPoint, None, newParent)
        newNode.children = [child.transformed(trans, newNode) for child in self.children]
        return newNode
    @staticmethod
    def makeLinkedList(points):
        head = None
        prev = None
        for p in points:
            nxt = TreeNode(p)
            if prev is None:
                head = nxt
            else:
                prev.addChild(nxt)
            prev = nxt
        return head
    def calcPolygonEdgesAndPoints(self, cap=False, capStart=False, reverseCaps=False,
                                  reverseStartCap=False):
        edges, points = [], []
        for child in self.children:
            edgeFirst = Line.fromPoints(*self.point, *child.point, segment=True)
            edgeLast = edgeFirst.reversed()
            edges.append(edgeFirst)
            points.append(self.point)
            if child.children:
                childEdges, childPoints = child.calcPolygonEdgesAndPoints(cap=cap,
                                                    reverseCaps=reverseCaps)
                edges.extend(childEdges)
                points.extend(childPoints)
            elif cap:
                capEdge = edgeFirst.makeCap(child.point)
                if reverseCaps:
                    capEdge.reverse()
                edges.append(capEdge)
                points.append(child.point)
            edges.append(edgeLast)
            points.append(child.point)
        if capStart and len(self.children) == 1:
            capEdge = edgeLast.makeCap(self.point)
            if reverseCaps ^ reverseStartCap:
                capEdge.reverse()
            edges.append(capEdge)
            points.append(self.point)
        return edges, points
    def toDrawables(self, elements, drawVerts=False, drawEdges=False,
                    drawContinuous=False, reverseCaps=False, reverseStartCap=False,
                    **kwargs):
        ds = []
        if drawContinuous:
            edges, points = self.calcPolygonEdgesAndPoints(cap=True, capStart=True,
                                        reverseCaps=reverseCaps,
                                        reverseStartCap=reverseStartCap)
            poly = Polygon(edges=edges, vertices=points, join=False)
            #poly.offsetPolygon()
            ds0 = poly.toDrawables(elements, **kwargs)
            ds.extend(ds0)
        if drawEdges:
            for child in self.children:
                edge = Line.fromPoints(*self.point, *child.point, segment=True)
                ds0 = edge.toDrawables(elements, **kwargs)
                ds.extend(ds0)
        if drawVerts:
            ds0 = self.point.toDrawables(elements, **kwargs)
            ds.extend(ds0)
        for child in self.children:
            ds.extend(child.toDrawables(elements, drawVerts=drawVerts, drawEdges=drawEdges,
                                        drawContinuous=False, **kwargs))
        return ds
```

```python
def makeHyperDragon(p1, p2, depth, reverse=False, angle=90, scale=1, fraction=0.5):
    if depth <= 0:
        head = TreeNode(p1)
        tail = TreeNode(p2)
        if reverse:
            head, tail = tail, head
        head.addChild(tail)
        return head, tail
    else:
        d = p1.distanceTo(p2)
        trans1 = Transform.translation(Point.fromHPolar(d*fraction,0))
        trans2 = Transform.translation(p1, p2)
        trans = Transform.merge(trans1, trans2)
        pMid = trans.applyToPoint(Point.fromHPolar(d/2*scale,deg=angle))
        head1, tail1 = makeHyperDragon(pMid, p1, depth-1, reverse=not reverse,
                                       angle=angle, scale=scale, fraction=fraction)
        head2, tail2 = makeHyperDragon(pMid, p2, depth-1, reverse=reverse,
                                       angle=angle, scale=scale, fraction=fraction)
        if reverse:
            head1, tail1, head2, tail2 = head2, tail2, head1, tail1
        tail1.addChild(head2.children[0])
        return head1, tail2

def genP1P2():
    global p1, p2
    p1 = Point.fromHPolar(length/2, deg=0)
    p2 = Point.fromHPolar(length/2, deg=180)

choice = 'mod1'
if choice == 'mod1':
    # Dragon mod1
    depth = 9
    length = 3
    genP1P2()
    dragon, _ = makeHyperDragon(p1, p2, depth, scale=-1/math.sqrt(3), fraction=0.4)
elif choice == 'mod2':
    # Dragon mod2 (tree, mountain)
    depth = 6
    length = 1
    genP1P2()
    dragon, _ = makeHyperDragon(p1, p2, depth, scale=.9, fraction=1.1)
else:
    assert False

d = Drawing(2.01, 2.01, origin='center')
d.draw(euclid.shapes.Circle(0, 0, 1),
       fill='none', stroke_width=0.006, stroke='black')
lineStyle = dict(hwidth=0.03, fill='black')
d.draw(dragon, drawContinuous=True,
       hwidth=(-0.002*3,-0.008*3), fill='black', opacity=0.7)

d.setRenderSize(w=600)
d.saveSvg('images/dragon_{}.svg'.format(choice))
d
```
